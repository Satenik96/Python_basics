```python
from nltk.util import ngrams
final_list=[]
def bi_gram(a):
    for i in range(0, len(a)):
        a_list=list(ngrams(a[i], 2))
        final_list.append(a_list)
    return final_list
```

```python
final_list = bi_gram(a)
final_list
```

```python
def smt(dataset):
    final_s = []
    final_s_temp =[]
    for i in range(0,len(dataset)):
        for j in range(0, len(dataset[i])):
            s1=dataset[i][j][0].lower()
            s2=dataset[i][j][1].lower()
            final_s_temp.append(s1+s2)
            if j==len(dataset[i])-1:
                final_s.append(final_s_temp)
                final_s_temp = []
    return final_s
```

```python
final_final_list = smt(final_list)
final_final_list
```

```python
import itertools
alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

from itertools import *

keywords = [''.join(i) for i in itertools.product(alphabets, repeat = 2)]
```

```python
#first 5 elements of keywords
keywords[0:5]
```

```python
def zero_one(dataset, alphabet):
    list_2 = []
    for i in range(0,len(dataset)):
        list_1=[]
        for l in alphabet:
            if l in dataset[i]:
                list_1.append(1)
            else:
                list_1.append(0)
        list_2.append(list_1)
    return list_2
```

```python
list_zero_one = zero_one(final_final_list, keywords)
```

```python
df = pd.DataFrame(list_zero_one, columns = keywords)
```

```python
df.head()
```

```python
df['label'] = b
```

```python
df.head()
```

```python
#imbalanced dataset
df['label'].value_counts()
```

```python
#balancing the data
one = len(df[df['label'] == 1])
one_indices = df[df.label == 0].index
random_indices = np.random.choice(one_indices,one, replace=False)
zero_indices = df[df.label == 1].index
under_sample_indices = np.concatenate([zero_indices,random_indices])
under_sample = df.loc[under_sample_indices]
```

```python
#already balanced
under_sample['label'].value_counts()
```

```python
#Logistic Regression
X_under = under_sample.loc[:,under_sample.columns != 'label']
y_under = under_sample.loc[:,under_sample.columns == 'label']
X_under_train, X_under_test, y_under_train, y_under_test = train_test_split(X_under,y_under,test_size = 0.3, random_state = 0)

lr_under = LogisticRegression(penalty='l1', tol=0.01, C=0.9)
clf = lr_under.fit(X_under_train,y_under_train)
y_under_pred = lr_under.predict(X_under_test)
```

```python
print('Recall is',recall_score(y_under_test,y_under_pred))
print('Accuracy is',accuracy_score(y_under_test,y_under_pred))
print('Precision is',precision_score(y_under_test,y_under_pred))
```
